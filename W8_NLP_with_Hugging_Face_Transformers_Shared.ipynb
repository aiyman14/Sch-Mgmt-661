***Name Entity Recognition (NER)***

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    model="tner/roberta-large-ontonotes5", # you can use the default or specify a model
    aggregation_strategy="simple"   # merges subword pieces
)

text = "Airbnb will open a hub in Paris on Jan 2025 with a $5M investment."
ner(text)
```

***Part of Speech (POS)***

```python

pos_tagger = pipeline(
    "token-classification",
    model="vblagoje/bert-english-uncased-finetuned-pos",
    aggregation_strategy="simple"
)

text = "The quick brown fox jumps over the lazy dog."
pos_tagger(text)
```

***Zero-shot Classification***

```python
from transformers import pipeline

clf = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"   # multilingual: joeddav/xlm-roberta-large-xnli
)

text = "I love hiking in the mountains on weekends."
candidate_labels = ["travel", "sports", "food", "finance"]

result = clf(text, candidate_labels, multi_label=False)
result
```

***Text Summarization***

```python
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"     # long docs: pszemraj/led-large-book-summary
)

text = ("""Deep learning is a specialized branch of machine learning that focuses on the use of artificial
neural networks with many layers to process and analyze large amounts of data. Unlike traditional machine l
earning methods, which often rely on handcrafted features and explicit rules, deep learning systems automatically
learn representations of data by adjusting millions of parameters within the network. This approach has proven
particularly effective in domains where patterns are complex and difficult to specify manually,
such as computer vision, speech recognition, natural language processing, and game playing""")

summary = summarizer(text, max_length=50, min_length=15)

summary[0]['summary_text']
```

```python
summary
```

***Translation***

```python
# Initialize translation pipeline (English → French example)
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")

text = "Deep learning is transforming artificial intelligence."
translation = translator(text)

print(translation[0]['translation_text'])
```

***Sentiments and Emotions***

```python

# Initialize sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

text = "The new product update is amazing and super useful!"
result = sentiment_analyzer(text)

print(result)

```

```python
# Initialize emotions pipeline

emotions = pipeline(
    task="text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=4,      # top emotion only
    truncation=True

)

texts = "I’m so excited about our trip!"
print(emotions(texts))




```

```python
from transformers import pipeline

clf = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"   # multilingual: joeddav/xlm-roberta-large-xnli
)

text = "I love hiking in the mountains on weekends."
candidate_labels = ["positive", "negative"]

result = clf(text, candidate_labels, multi_label=False)
result
```
